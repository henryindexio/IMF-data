```python
# This file graphs the country bond average yields and the number of defaults data from BondYieldsData.ipynb.
```

```python
import pickle

f = open('MeanGroupYields.pckl', 'rb')
MeanGroupYields = pickle.load(f)
f.close()

f2 = open('NDefaults.pckl', 'rb')
NDefaults = pickle.load(f2)
f2.close()

f3 = open('RatingLabels.pckl', 'rb')
RatingLabels = pickle.load(f3)
f3.close()
```

```python
import matplotlib.pyplot as plt
import numpy as np

N = 7
ind = np.arange(N)
width = 0.35 

fig,ax = plt.subplots()
YieldBar = ax.bar(ind, MeanGroupYields, width, color='g')

ax.set_ylabel('Yield(%)')
ax.set_xlabel('Rating Group')
ax.set_title('Mean Country Bond Yields by Rating')
ax.set_xticklabels(['0','AAA','AA','A','BBB','BB','B','C'])

plt.show()
```

```python
N = 7
ind = np.arange(N)
width = 0.35 

fig,ax = plt.subplots()
YieldBar = ax.bar(ind, NDefaults, width, color='k')

ax.set_ylabel('Defaults')
ax.set_xlabel('Rating Group')
ax.set_title('Number of Country Bond Defaults by Rating')
ax.set_xticklabels(['0','AAA','AA','A','BBB','BB','B','C'])

plt.show()
```
